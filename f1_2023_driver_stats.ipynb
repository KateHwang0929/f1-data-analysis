# Section 1: Dataset

For this project I chose a dataset about the 2023 Formula 1 World Championship driver season. Each row in my dataset represents one F1 driver who took part in the 2023 season, and each column stores a data that summarizes their performance across the whole year. The columns include basic information such as the driver’s name, nationality, and team, as well as performance statistics like total points, wins, podiums, pole positions, fastest laps, number of races where they scored points, and how many times they did not finish a race (DNFs).

I built this dataset by combining the official 2023 F1 Drivers’ Standings from the Formula 1 website with more detailed statistics from GP Racing stats, and saving everything in one CSV file called `f1_2023_driver_stats.csv`. It is meaningful to me because I am very interested in Formula 1 and I fell in love with Formula 1 since 2023. Thus it is fun to connect what I watched with data. I am especially curious about how consistency and reliability (for example, points finishes and DNFs) relate to total points and overall success.

The dataset has 22 rows (one for each driver) and 12 columns, which gives 264 data points. This is above the required 200 data points. Some of the key columns are:

- `points`: total championship points scored in 2023  
- `wins`: number of race wins  
- `podiums`: number of times the driver finished in the top 3 (who were able to stand on the podium!)
- `poles`: number of pole positions in qualifying  
- `fastest_laps`: number of fastest laps in races  
- `points_finishes`: number of races where the driver scored points  
- `points_per_points_finish`: average points in races where they scored  
- `dnfs`: number of races the driver did not finish


```python
import pandas as pd

f1 = pd.read_csv("f1_2023_driver_stats.csv")
f1
```

# Section 2: Exploratory Data Analysis

In this section I use descriptive statistics to get a broad overview of the 2023 F1 driver performances. For quantitative variables like `points`, `wins`, `podiums`, and `dnfs`, useful descriptive statistics include:

- **Measures of center**: mean and median points help describe a “typical” season result.
- **Measures of spread**: standard deviation and the minimum/maximum values show how unequal the distribution is (for example, how much Max Verstappen dominates compared to drivers at the back).
- **Counts**: the `points_finishes` and `dnfs` columns tell me how often drivers finished in the points versus how often they retired.

I also look at sorted tables to compare drivers. For example, sorting by `points` gives the championship order, while sorting by `dnfs` shows which drivers had the most retirements. These simple statistics already start to tell a story about which drivers were consistently strong, which ones were fast but unlucky, and which ones struggled.


```python

numeric_cols = ["points", "wins", "podiums", "poles",
                "fastest_laps", "points_finishes", "points_per_points_finish", "dnfs"]


f1_points_sorted = f1.sort_values("points", ascending=False)
f1_points_sorted[["driver", "team", "points", "wins", "podiums"]]
f1_dnfs_sorted = f1.sort_values("dnfs", ascending=False)
f1_dnfs_sorted[["driver", "team", "points", "dnfs"]]
```

Looking specifically at the number of DNFs (Did Not Finish) among the 2023 F1 drivers, the descriptive statistics show a clear picture of how reliability varied across the grid. The average driver had about 2.77 DNFs, while the median was slightly lower at 2.5, meaning half of the drivers retired from two or fewer races and half retired from more. The minimum value was 0, indicating that a few drivers completed every race without retiring, while the maximum was 7 DNFs, showing that some drivers faced frequent reliability issues or incidents. The spread between 0 and 7 DNFs highlights how inconsistent reliability was across the field, and it also helps explain why drivers with higher DNF counts tended to score fewer total points during the season.

# Section 5: Data Science

For Part II, I chose to focus on the question: **How strongly do DNFs predict a driver’s total points in the 2023 F1 season?** In Section 4, I observed that drivers with fewer DNFs generally scored more points, and I want to measure this relationship more formally. Thus, I will use a simple linear regression model (that we learned during the lab and lectures). The model will predict a driver’s total points using only their number of DNFs as the input variable.

The goal is to see whether DNFs alone can explain a meaningful amount of the variation in points. In other words, if DNFs are a strong predictor, the model should have a clear negative slope. If not, it will show that reliability matters but cannot fully explain differences in performance between drivers. 

```python
from sklearn.linear_model import LinearRegression
import numpy as np

X = f1[["dnfs"]]       
y = f1["points"]       

model = LinearRegression()
model.fit(X, y)

#applied from the lab we did.
slope = model.coef_[0]
intercept = model.intercept_

y_pred = model.predict(X)

slope
```

```python
intercept
```

**Interpretation:** The linear regression model shows a negative slope, meaning that as DNFs increase, a driver’s predicted total points decrease. The intercept of 182.6 means that  model predicts a driver with 0 DNFs would score approximately 182 points.
